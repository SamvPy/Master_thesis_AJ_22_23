```python
from __future__ import division
import pandas as pd
import numpy as np
import sqlite3
import mysql.connector
import csv
```

```python
conn = mysql.connector.connect(user='root', password='password', host='127.0.0.1', port='3306',
                               database='expression_atlas_cells2')
mycursor = conn.cursor()

# check the connection
if conn.is_connected():
    print("connection succesfull")
else:
    print("no connection")
```

```python
uniprot = pd.read_csv('Uniprot.csv', sep=',')
```

```python
uniprot
```

```python
seq = pd.read_csv('Uniprot_sequences.csv', sep=';')
seq[seq['uniprot_id'].duplicated() == True]
```

```python
uni_seq = pd.merge(uniprot, seq, on='uniprot_id', how='left')
uni_seq = uni_seq.drop_duplicates()
```

```python
uni_seq
```

```python
mysql = "SELECT DISTINCT(uniprot_id) from protein"
mysqlData = pd.read_sql_query(mysql, conn)
mysqlData.head()
```

```python
mysqlData['description'] = np.nan
mysqlData['length'] = np.nan
mysqlData['sequence'] = np.nan
```

```python
mysqlData
```

```python
total_ids = pd.merge(uni_seq, mysqlData,how='right', on='uniprot_id')
```

```python
total_ids.shape
```

```python
total_ids.head()
```

```python
total_ids = total_ids.drop(columns=['description_y', 'length_y', 'sequence_y'])
total_ids = total_ids.rename(columns={'description_x' : 'description', 'length_x' : 'length', 'sequence_x': 'sequence'})
```

```python
total_ids.to_csv('protein_table_uniprot_updated.csv', sep=',', index=False)
```

```python
uniprot_csv = csv.reader(open('protein_table_uniprot_updated.csv'), delimiter=',')
for row in uniprot_csv:
    uni_insert = "INSERT INTO protein(uniprot_id, description, length, sequence) VALUES(%s, %s, %s, %s)"
    mycursor.execute(uni_insert, row) 
    conn.commit()
print('Done')
```

```python
# gene ontology
go_csv = csv.reader(open('gene_ontology.csv'), delimiter=',')
for row in go_csv:
    go_csv = "INSERT INTO gene_ontology(go_id, go_term, go_domain) VALUES(%s, %s, %s)"
    mycursor.execute(go_csv, row) 
    conn.commit()
print('Done')
```

```python
#All proteins in the protein_to_go file should be found in the original protein table. Otherwise, an error will appear
prot_go = pd.read_csv('protein_to_go.csv', sep=',')
proteins = pd.read_csv('protein_table_uniprot_updated.csv', sep=',')
gos = pd.read_csv('gene_ontology.csv', sep=',')
proteins = proteins['uniprot_id'].tolist()
prot_go = prot_go[prot_go['uniprot_id'].isin(proteins)]
prot_go = prot_go[prot_go['go_term'].isin(gos)]
```

```python
prot_go.to_csv('protein_to_go_updated.csv', sep=',', index=False)
```

```python
# protein to go
# tissue to protein
prot_go_csv = csv.reader(open('protein_to_go_updated.csv'), delimiter=',')
for row in prot_go_csv:
    disable = "SET FOREIGN_KEY_CHECKS=0"
    mycursor.execute(disable) 
    prot_go_insert = "INSERT INTO protein_to_go(uniprot_id, go_id) VALUES(%s,%s) ON DUPLICATE KEY UPDATE uniprot_id=uniprot_id, go_id=go_id"
    mycursor.execute(prot_go_insert, row) 
    conn.commit()
print('Done')
```
